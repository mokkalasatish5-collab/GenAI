```python

from langchain_core.prompts import ChatPromptTemplate


# ---------------------
# 2. Function: Extract a role from input text
# ---------------------
def extract_role(user_text):
    roles = ["data scientist", "ml engineer", "ai engineer",
             "web developer", "full stack developer", "gen ai engineer"]

    text_lower = user_text.lower()
    matches = [r for r in roles if r in text_lower]

    # If multiple roles found → choose the first
    if matches:
        return matches[0]

    # If no role found → default role
    return "software engineer"


# ---------------------
# 3. Prompt Template with a variable {role}
# ---------------------
prompt_template = ChatPromptTemplate.from_messages([
    ("system",
     "You are an expert interview coach. Answer only for the role: {role}."),
    ("user",
     "I want to prepare for the role {role}. Give me guidance.")
])

# ---------------------
# 4. Full chain
# ---------------------
def interview_coach(user_input):
    # Step A: extract role
    role = extract_role(user_input)
    print(f"Detected role → {role}")

    # Step B: format prompt
    final_prompt = prompt_template.format(role=role)

    # Step C: run LLM
    result = llm.invoke(final_prompt)
    return result.content


# ---------------------
# 5. Test the workflow
# ---------------------
user_text = "I want to prepare for interviews in AI Engineer, ML Engineer and Data Scientist."
response = interview_coach(user_text)
print("\n--- LLM Response ---")
print(response)
```
